Convert Data To Array

```python
data.columns
```

```python
x = data[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh',
       'exng', 'oldpeak', 'slp', 'caa', 'thall']].values
x[0:10]
```

```python
y = data['output'].values
y[0:10]
```

Normalize The Data
تابع نرمال

```python
Scaler = preprocessing.StandardScaler().fit(x)
X = Scaler.transform(x.astype(float))
print(X[0:5])
```

کد بالا داده های ما را روی نمودار نرمال قرار داد

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
print (X_train.shape, Y_train.shape)
```

```python
print (X_test.shape, Y_test.shape)
```

Classification

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k = 4
neigh = KNeighborsClassifier(n_neighbors= k).fit(X_train,Y_train)
neigh
```

Testing The Model

```python
yHat = neigh.predict(X_test)
print(y[0:5])
print(yHat[0:5])
```

Accuracy Evaluation

```python
from sklearn import metrics
```

```python
Train_Set_Accuracy = metrics.accuracy_score(Y_train,neigh.predict(X_train))
Test_Set_Accuracy = metrics.accuracy_score(Y_test,yHat)
print(Train_Set_Accuracy)
print(Test_Set_Accuracy)
```

Find The Best K

```python
KS = 10
mean_acc = np.zeros((KS-1))
std_acc = np.zeros((KS-1))

for i in range(1,KS):
    neigh = KNeighborsClassifier(n_neighbors = i).fit(X_train,Y_train)
    yHat = neigh.predict(X_test)
    mean_acc[i-1] = metrics.accuracy_score(Y_test,yHat)
    std_acc[i-1] = np.std(yHat==Y_test)/np.sqrt(yHat.shape[0])

mean_acc
```

Draw Chart

```python
plt.plot(range(1,KS),mean_acc,'g')
plt.fill_between(range(1,KS),mean_acc - 1 * std_acc , mean_acc + 1 * std_acc , alpha = 0.10)
plt.fill_between(range(1,KS),mean_acc - 3 * std_acc , mean_acc + 3 * std_acc , alpha = 0.10, color = 'g')
plt.legend(('Accuracy', '+/- x std' , '+/- 3x std'))
plt.ylabel('Accuracy')
plt.xlabel('Numbers of k')
plt.tight_layout()
plt.show()
```

طبق نتایج بالا بهترین مقدار k = 3 است اما با این حال مدل خوبیس ساخته نشد
